```python
# Finding the IQR
percentile25 = df['Interdomain spacing (nm)'].quantile(0.25)
percentile75 = df['Interdomain spacing (nm)'].quantile(0.75)
```

```python
percentile25
```

```python
iqr = percentile75 - percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
upper_limit,lower_limit
```

```python
# finding outlier
df[df['Interdomain spacing (nm)'] > upper_limit]
```

```python
df[df['DACH(%)'] < lower_limit]
```

```python
# Capping
new_df_cap = df.copy()

new_df_cap['Interdomain spacing (nm)'] = np.where(
    new_df_cap['Interdomain spacing (nm)'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['Interdomain spacing (nm)'] < lower_limit,
        lower_limit,
        new_df_cap['Interdomain spacing (nm)']
    )
)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

# Prediction for Phase Seperation



```python
X= df.drop(columns=['Phase Separation(%)','Copolymer Identifier'])
```

```python
X
```

```python
y = df[['Phase Separation(%)']]
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Applying Linear Regression
```

```python
## scaling the column
scaler= StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(scaler,lr)
pipe.fit(X_train,y_train)
```

```python
y_pred_lr= pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
## Lasso Regression
```

```python
lasso= Lasso()
```

```python
pipe = make_pipeline(scaler,lasso)
pipe.fit(X_train,y_train)
```

```python
y_pred_lasso= pipe.predict(X_test)
r2_score(y_test,y_pred_lasso)
```

```python
# Ridge
```

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(scaler,ridge)
pipe.fit(X_train,y_train)
```

```python
y_pred_ridge= pipe.predict(X_test)
r2_score(y_test,y_pred_ridge)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
DT = DecisionTreeClassifier(random_state=2)
```

```python
pipe = make_pipeline(scaler,DT)
pipe.fit(X_train,y_train)
```

```python
y_pred_DT= pipe.predict(X_test)
r2_score(y_test,y_pred_DT)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
RF =  RandomForestRegressor(random_state=2,n_estimators=2)
```

```python
pipe = make_pipeline(scaler,RF)
pipe.fit(X_train,y_train)
```

```python
y_pred_RF= pipe.predict(X_test)
r2_score(y_test,y_pred_RF)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.9, random_state=42)
```

```python
pipe = make_pipeline(scaler,elastic_net)
pipe.fit(X_train,y_train)
```

```python
y_pred_EN= pipe.predict(X_test)
r2_score(y_test,y_pred_EN)
```
